```python
import pandas as pd
```

```python
df=pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
df = df[['text', 'label_num']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label_num'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
mnb = MultinomialNB()
mnb.fit(X_train_counts, y_train)
y_pred_mnb = mnb.predict(X_test_counts)
```

```python
print("MultinomialNB:")
print("Accuracy:", accuracy_score(y_test, y_pred_mnb))
print("Report:\n", classification_report(y_test, y_pred_mnb))
```

```python
gnb = GaussianNB()
X_train_dense = X_train_counts.toarray()
X_test_dense = X_test_counts.toarray()
gnb.fit(X_train_dense, y_train)
y_pred_gnb = gnb.predict(X_test_dense)
```

```python
print("\nGaussianNB:")
print("Accuracy:", accuracy_score(y_test, y_pred_gnb))
print("Report:\n", classification_report(y_test, y_pred_gnb))
```

```python
sample_messages = [
    "Congratulations! You've won a free iPhone. Click to claim.",
    "Hey, are we still meeting tomorrow for the project?",
    "Get cheap meds now! Visit our website for great deals.",
    "Hi Mom, just reached college. Talk to you later.",
    "Win a brand new car by entering this lucky draw!"
]
```

```python
sample_vectors = vectorizer.transform(sample_messages)

# Make predictions
predictions = mnb.predict(sample_vectors)

# Map numeric labels back to string labels (if needed)
label_map = {0: 'ham', 1: 'spam'}
predicted_labels = [label_map[p] for p in predictions]

# Print results
for msg, label in zip(sample_messages, predicted_labels):
    print(f"\nMessage: {msg}\nPredicted Label: {label}")

```
