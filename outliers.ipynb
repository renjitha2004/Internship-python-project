```python
import pandas as pd
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head()
```

```python
df=df.dropna()
```

```python
Q1=df['price'].quantile(0.25)
Q3=df['price'].quantile(0.75)
IQR=Q3-Q1
```

```python
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
```

```python
df_no_outliers=df[(df['price']>=lower_bound)&(df['price']<=upper_bound)]
```

```python
print("original dataset shape:",df.shape)
```

```python
print("After removing outliers:",df_no_outliers.shape)
```

```python
mean=df['price'].mean()
```

```python
std=df['price'].std()
```

```python
lower=mean-3*std
upper=mean+3*std
df_mean_filtered=df[(df['price']>=lower)&(df['price']<=upper)]
```

```python
print("After filtering out:",df_mean_filtered.shape)
```

```python
import numpy as np
```

```python
from statsmodels.robust.scale import mad
```

```python
median=df['price'].median()
mad_value=mad(df['price'])
threshold=3
df_mad_filtered=df[(np.abs(df['price']-median)/mad_value)<threshold]
```

```python
print ("After mad filtering",df_mad_filtered.shape)
```

```python
lower=df['price'].quantile(0.01)
upper=df['price'].quantile(0.99)
df_pct_filtered=df[(df['price']>=lower)&(df['price']<=upper)]
```

```python
print("After percentile filtering:",df_pct_filtered.shape)
```
